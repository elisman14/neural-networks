## Лабораторная работа №1 (Многоклассовая классификация цветов)

```python
import pandas
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

Считывание и разделение данных на входные и выходные

```python
dataframe = pandas.read_csv("iris.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
```

Преобразование данных из вектора в матрицу 0 и 1

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_y = to_categorical(encoded_Y)
```

Задание базовой архитектуры НС (3 слоя) 

```python
model = Sequential()
model.add(Dense(4, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))
```

Инициализация параметров обучения

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

Обучение ИНС

```python
history = model.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
```

Создание ИНС с другой архитектурой (увеличение количества нейронов на скрытом слое)

```python
model1 = Sequential()
model1.add(Dense(25, activation = 'relu'))
model1.add(Dense(3, activation = 'softmax'))
model1.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

Обучение второй модели ИНС

```python
history1 = model1.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
```

Создание третьей ИНС (увеличение количества скрытого слоя до 2)

```python
model2 = Sequential()
model2.add(Dense(25, activation = 'relu'))
model2.add(Dense(20, activation = 'relu'))
model2.add(Dense(3, activation = 'softmax'))
model2.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

Обучение третьей модели ИНС

```python
history2 = model2.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
```

Построение графиков потерь для этих моделей

```python
plt.plot(history.history['loss'], color = 'r', label = 'model')
plt.plot(history1.history['loss'], color = 'b', label = 'model1')
plt.plot(history2.history['loss'], color = 'g', label = 'model2')
plt.grid(True)
plt.legend()
plt.show()
```

,,Построение графиков точности для этих моделей

```python
plt.plot(history.history['accuracy'], color = 'r', label = 'model')
plt.plot(history1.history['accuracy'], color = 'b', label = 'model1')
plt.plot(history2.history['accuracy'], color = 'g', label = 'model2')
plt.grid(True)
plt.legend()
plt.show()
```

Обучение первой модели с другими параметрами (первая версия)

```python
model3 = Sequential()
model3.add(Dense(4, activation = 'relu'))
model3.add(Dense(3, activation = 'softmax'))
model3.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
history3 = model3.fit(X, dummy_y, epochs=40, batch_size=20, validation_split=0.2)
```

Обучение первой модели с другими параметрами (вторая версия)

```python
model4 = Sequential()
model4.add(Dense(4, activation = 'relu'))
model4.add(Dense(3, activation = 'softmax'))
model4.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
history4 = model4.fit(X, dummy_y, epochs=100, batch_size=15, validation_split=0.2)
```

Построение графиков потерь для моделей с одинаковой архитектурой и разными параметрами обучения

```python
plt.plot(history.history['loss'], color = 'r', label = 'model')
plt.plot(history3.history['loss'], color = 'b', label = 'model3')
plt.plot(history4.history['loss'], color = 'g', label = 'model4')
plt.grid(True)
plt.legend()
plt.show()
```

Построение графиков точности для моделей с одинаковой архитектурой и разными параметрами обучения

```python
plt.plot(history.history['accuracy'], color = 'r', label = 'model')
plt.plot(history3.history['accuracy'], color = 'b', label = 'model3')
plt.plot(history4.history['accuracy'], color = 'g', label = 'model4')
plt.grid(True)
plt.legend()
plt.show()
```

Вывод: наилучшая модель оказалось с наибольшим количеством слоев
