## Лабораторная работа №2 (Бинарная классификация отраженных сигналов радара)

```python
import pandas
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

Считывание и разделение данных на входные и выходные

```python
dataframe = pandas.read_csv("sonar.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
```

Преобразование данных в вектор 0 и 1

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

Задание базовой архитектуры НС (3 слоя)

```python
model = Sequential()
model.add(Dense(60, input_dim = 60, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))
```

Инициализация параметров обучения

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

Обучение ИНС

```python
history = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

Считывание и разделение данных на входные и выходные, но входных парамтеров в два раза меньше

```python
dataframe = pandas.read_csv("sonar.csv", header=None)
dataset = dataframe.values
X1 = dataset[:,0:30].astype(float)
Y1 = dataset[:,60]
```

```python
encoder1 = LabelEncoder()
encoder1.fit(Y1)
encoded_Y1 = encoder1.transform(Y1)
```

Задание базовой архитектуры ИНС и инициализация параметров обучения

```python
model1 = Sequential()
model1.add(Dense(60, input_dim = 30, kernel_initializer = 'normal', activation = 'relu'))
model1.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))
model1.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

Обучение второй ИНС

```python
history1 = model1.fit(X1, encoded_Y1, epochs=100, batch_size=10, validation_split=0.1)
```

Задание базовой архитектуры ИНС и инициализация параметров обучения для 3 модели

```python
model2 = Sequential()
model2.add(Dense(60, input_dim = 60, kernel_initializer = 'normal', activation = 'relu'))
model2.add(Dense(15, kernel_initializer = 'normal', activation = 'relu'))
model2.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))
model2.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

Обучение третьей ИНС

```python
history2 = model2.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

Задание базовой архитектуры ИНС и инициализация параметров обучения для 4 модели

```python
model3 = Sequential()
model3.add(Dense(60, input_dim = 60, kernel_initializer = 'normal', activation = 'relu'))
model3.add(Dense(25, kernel_initializer = 'normal', activation = 'relu'))
model3.add(Dense(15, kernel_initializer = 'normal', activation = 'relu'))
model3.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))
model3.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

Обучение четвертой ИНС

```python
history3 = model3.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

Построение графиков точности для этих моделей
